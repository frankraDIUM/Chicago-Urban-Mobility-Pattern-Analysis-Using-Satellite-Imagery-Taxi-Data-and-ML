```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile correctly
boundary_path = r'I:\GEO DATA ANALYSIS\chicago_mobility\Chicago_city.shp'

chicago_boundary = gpd.read_file(boundary_path)

# Check original CRS
print("Original CRS:", chicago_boundary.crs)

# Reproject to WGS84 (EPSG:4326) 
chicago_boundary = chicago_boundary.to_crs("EPSG:4326")

print("Reprojected CRS:", chicago_boundary.crs)

# Plot
fig, ax = plt.subplots(figsize=(10, 8))
chicago_boundary.plot(ax=ax, color='lightblue', edgecolor='black')
ax.set_title("Chicago City Boundary")
plt.show()

# Save
chicago_boundary.to_file("chicago_boundary_wgs84.geojson", driver="GeoJSON")
```
